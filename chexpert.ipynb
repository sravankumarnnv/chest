```python
import pandas as pd
import matplotlib.pyplot as plt
%pip install seaborn
import seaborn as sns
import os
```

```python
# Load the dataset
file_path = "/Volumes/2TB/chest/CheXpert_Small/train.csv"
df = pd.read_csv(file_path)

# Display the first few rows
print("Dataset preview:")
df.head()
```

```python
# Get all column names that represent medical findings/labels (excluding metadata columns)
label_columns = df.columns[5:]  # Skip the first 5 columns which are metadata (Path, Sex, Age, etc.)

# Display all unique label names
print("Medical finding labels in the CheXpert dataset:")
for label in label_columns:
    print(f"- {label}")
    
# Count how many labels we have
print(f"\nTotal number of medical finding labels: {len(label_columns)}")

# Optionally, show unique values for each label to understand the coding system
print("\nUnique values for each label (showing the CheXpert labeling system):")
for label in label_columns:
    unique_values = df[label].unique()
    print(f"{label}: {unique_values}")
```
